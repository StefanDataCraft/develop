```python
import re
regex = r"(Hallo|Hi|Hej) \w+, \1 \w+" #\1 repeats capturing group, must be repeated
m = re.search(regex, "Hallo Alice, Hallo Bob")
print(m.groups())
m = re.search(regex, "Hallo Alice, Hi Bob")
print(m.groups())
```

```python
fraud_mail = open("fraud_email.csv","r",encoding="ansi")
inhalt = fraud_mail.read()
emails = []
fraud_mail.close()
email_pattern = r"@\w{2,20}\.\w{2}"
emails = re.findall(email_pattern,inhalt, re.I)
emails = [email[-2:] for email in emails]
emails

email_counts = {country:emails.count(country)for country in set(emails)}
email_counts

import pandas as pd
df = pd.Series(email_counts)
df.plot.pie()
```
